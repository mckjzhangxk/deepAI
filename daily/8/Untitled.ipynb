```python
import numpy as np
import matplotlib.pyplot as plt
from keras.layers import Dense,ReLU,Input
from keras.models import Model
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
from keras.losses import binary_crossentropy
```

# First make data set

```python
def dataset(m1,m2):
    w1,b1=5,8
    w2,b2=1,-4
    
    t1=np.arange(m1)
    y1=w1*t1+b1+np.random.randn(m1)*5
    d1=np.stack([t1,y1],axis=1)
    
    t2=np.arange(m2)
    y2=w2*t2+b2+np.random.randn(m2)*4
    d2=np.stack([t2,y2],axis=1)
    
    return d1,d2
```

```python
d1,d2=dataset(40,30)
plt.scatter(d1[:,0],d1[:,1])
plt.scatter(d2[:,0],d2[:,1])
```

```python
X=np.concatenate([d1,d2],axis=0)
Y=np.concatenate([np.ones(d1.shape[0]),np.zeros(d2.shape[0])],axis=0)
```

# Define model

```python
def MyLinearModel():
    X_input=Input(shape=(2,))
    X=Dense(1,activation='sigmoid')(X_input)
    
    model=Model(inputs=X_input,outputs=X)
    model.compile(optimizer='adam',loss=binary_crossentropy,metrics=['accuracy'])
    return model
```

```python
model=MyLinearModel()
SVG(model_to_dot(model).create(prog='dot', format='svg'))
```

```python
model.fit(X,Y,batch_size=10,epochs=1000)
```

```python
n=100
xrange=np.linspace(0,500,n)
yrange=np.linspace(500,0,n)
xg,yg=np.meshgrid(xrange,yrange)

c=model.predict(np.stack([xg.ravel(),yg.ravel()],axis=1)).reshape(n,n)
plt.contourf(xg,yg,c)
plt.colorbar()
plt.scatter(d1[:,0],d1[:,1])
plt.scatter(d2[:,0],d2[:,1])
```

```python
def datasetRegress(m1):
    w1,b1=5,8

    ret=[]
    for i in range(30,40):
        t1=np.arange(m1)
        y1=w1*t1+b1+np.random.randn(m1)*i
        d1=np.stack([t1,y1],axis=1)
        ret.append(d1)
    return np.concatenate(ret,axis=0)
```

```python
Dr=datasetRegress(100)
plt.scatter(Dr[:,0],Dr[:,1],marker='x',s=1)
```

```python
def MyRegModel():
    X_input=Input(shape=(1,))
    X=Dense(1,)(X_input)
    
    model=Model(inputs=X_input,outputs=X)
    
    model.compile(optimizer='adam',loss='mse')
    return model
```

```python
model=MyRegModel()
```

```python
model.fit(Dr[:,0],Dr[:,1],batch_size=100,epochs=2000)
```

```python
Dr=datasetRegress(100)
plt.scatter(Dr[:,0],Dr[:,1],marker='x',s=1)

plt.scatter(Dr[:,0],model.predict(Dr[:,0]),marker='.',s=1)
```
