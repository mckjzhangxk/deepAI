```python
import matplotlib.pyplot as plt
import numpy as np
import itertools
import os
import pickle
import face_recognition
from scipy.misc import imread,imresize,imshow
import pandas as pd
%matplotlib inline
```

```python
def gaussMeasure(x,u=1,w=1):return np.exp(-((x-u)**2)/w)
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

#     print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else '.2f'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Image X')
    plt.xlabel('Image Y')
    
mnist_data_path='../AI_database/mnist/MNIST_DATA'
```

```python
class FaceService():
    def __init__(self,N=100,D=128):
        self.N=N
        self.D=D
        self._db=np.zeros((N,D))
        self._num=0 #how many users have be registered
        self.faceid=[]

    def getNumOfFace(self):
        return self._num
    def loaddb(self,dir):
        N = 0
        for f in os.listdir(dir):
            N += 1

        for f in os.listdir(dir):
            self.addFace2db(os.path.join(dir,f))
            if self._num%50==0:
                print('load progress %f'%(self._num/N*100))
        with open('faces.pickle', 'wb') as handle:
            pickle.dump(self, handle, protocol=pickle.HIGHEST_PROTOCOL)
        print('load face database successful,load #%d faces!'%(self._num))
        print('=========================================================')
    def addFace2db(self,filename):
        image = self._handleImg(face_recognition.load_image_file(filename))
        characters = face_recognition.face_encodings(image)
        if len(characters)==0:return False
        self._appendFace(characters[0])
        self.faceid.append(filename[filename.rfind('/')+1:])
        self._num += 1
        return True
    def _appendFace(self,c):
        N=self._db.shape[0]
        if self._num>=N:
            _db = np.zeros((self.N, self.D))
            self._db=np.concatenate((self._db,_db),0)
        self._db[self._num]=c
    def _handleImg(self,img,scale=0.6):
        H,W,_=img.shape

        img=imresize(img,(int(scale*H),int(scale*W)))

        return img
    def detectFace(self,filename):
        img=face_recognition.load_image_file(filename)
        
        cf=face_recognition.face_encodings(img)
        if(len(cf)==0):return (None,1000)

        # distance=np.sum((self._db[:self._num]-cf[0])**2,1)
        # chioce=np.argmin(distance)
        distance=self._db[:self._num].dot(cf[0])
        norma=np.linalg.norm(self._db[:self._num],axis=1)
        normb=np.linalg.norm(cf[0])
        distance=distance/norma/normb

        chioce = np.argmax(distance)
        return (self.faceid[chioce],distance[chioce])
    
    def getEncodes(self):
        return self._db[:self._num]
```
