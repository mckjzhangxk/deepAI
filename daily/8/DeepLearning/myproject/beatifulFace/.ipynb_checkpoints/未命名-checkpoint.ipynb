```python
import numpy as np
import imutils
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# imagePaths=[
#         '1.jpeg','2.jpeg','3.jpeg'
# ]

imagePaths=[
        'd.jpeg','c.jpeg','b.jpeg','a.jpeg'
]
imgs=[cv2.imread(p) for p in imagePaths]
```

```python
imutils.find_function('aff')
```

```python
img=cv2.imread('car.png')
plt.imshow(img)
```

```python
tx,ty=30,80
T=np.array([[1.0,0.0,tx],[0.0,1.0,ty]])
img1=cv2.warpAffine(img,T,(img.shape[1],img.shape[0]))
plt.imshow(img1)
```

```python
img1=imutils.translate(img,tx,ty)
plt.imshow(img1)
```

```python
plt.imshow(imutils.rotate(img,90))
```

```python
R=cv2.getRotationMatrix2D((img.shape[1]//2,img.shape[0]//2),90,1)
plt.imshow(cv2.warpAffine(img,R,(img.shape[1],img.shape[0])))
```

```python
stitcher=cv2.Stitcher_create()
```

```python
status,pano=stitcher.stitch(imgs)
```

```python
plt.imshow(pano)
```

```python
pano_copy=cv2.copyMakeBorder(pano,10,10,10,10,cv2.BORDER_CONSTANT,(0,0,0))
pano_gray=cv2.cvtColor(pano_copy,cv2.COLOR_BGR2GRAY)
_,pano_mask=cv2.threshold(pano_gray,0,255,cv2.THRESH_BINARY)
plt.imshow(pano_mask,'gray')
```

```python
cnts = cv2.findContours(pano_mask.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
```

```python
c=max(cnts,key=cv2.contourArea)
(x,y,w,h)=cv2.boundingRect(c)

pano_mask_1=pano_mask.copy()
cv2.rectangle(pano_mask_1,(x,y),(x+w,y+h),255,-1)
plt.imshow(pano_mask_1,'gray')
```

```python
		# create two copies of the mask: one to serve as our actual
		# minimum rectangular region and another to serve as a counter
		# for how many pixels need to be removed to form the minimum
		# rectangular region
minRect = pano_mask_1.copy()
sub = pano_mask_1.copy()
		# keep looping until there are no non-zero pixels left in the
		# subtracted image
while cv2.countNonZero(sub) > 0:
			# erode the minimum rectangular mask and then subtract
			# the thresholded image from the minimum rectangular mask
			# so we can count if there are any non-zero pixels left
	minRect = cv2.erode(minRect, None)
	sub = cv2.subtract(minRect, pano_mask)#0-255=0
```

```python
cnts=imutils.grab_contours(cv2.findContours(minRect, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE))
c=max(cnts,key=cv2.contourArea)
(x,y,w,h)=cv2.boundingRect(c)
pano_final=pano_copy[y:y+h,x:x+w]
plt.imshow(pano_final,'gray')
```
