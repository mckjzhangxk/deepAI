```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
N=10000
mean,std=1,1
data=np.random.randn(N)*std+mean
plt.figure()
plt.hist(data)
plt.figure()
plt.hist(sigmoid(data))
```
