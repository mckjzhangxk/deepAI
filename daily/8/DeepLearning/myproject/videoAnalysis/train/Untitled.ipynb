```python
from matplotlib import pyplot as plt
import numpy as np
import os
import cv2
%matplotlib inline
```

```python
def readDataLabel(filepath,imgprefix):
    with open(filepath,'r') as fs:
        lines=fs.readlines()

    arr=[]
    face2Image={}
    multi=[]
    for line in lines[1:]:
        record=line.strip().split('\t')
        

    
        orgin=record[0]
        filename=record[1]

        faceid=record[2]
        age=str(record[3])
        gender=str(record[4])
        x,y,dx,dy=record[5:9]
        fname=os.path.join(imgprefix,orgin,'coarse_tilt_aligned_face.'+faceid+'.'+filename)

        if filename in face2Image:
            multi.append(fname)
        else:
            face2Image[filename]=1
            
        assert os.path.exists(fname),'%s,not exist'%(fname)      
#         if age==-1 and gender==-1:continue
        
        arr.append((fname,age,gender,(int(x),int(y),int(dx),int(dy))))
        
    for d in multi:
        print(d)
    print(len(d))
    return arr
```

```python
db=readDataLabel(
    filepath='../data/Adience/fold_0_data.txt',
    imgprefix='/home/zhangxk/AIProject/数据集与模型/faces')
```
