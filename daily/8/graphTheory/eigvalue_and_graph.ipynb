```python
import matplotlib.pyplot as plt
import networkx as nx
from scipy.linalg import eigh
%matplotlib inline
# https://www.johndcook.com/blog/2016/01/09/spectra-of-complete-graphs-stars-and-rings/
```

```python
G=nx.complete_graph(5)
nx.draw(G)
L=nx.laplacian_matrix(G)
S,V=eigh(L.toarray())
print(S)
print(L.toarray())
```

```python
G=nx.star_graph(5)
nx.draw(G)
L=nx.laplacian_matrix(G)
S,V=eigh(L.toarray())
print(S)
print(L.toarray())
```

```python
G=nx.cycle_graph(5)
nx.draw(G)
L=nx.laplacian_matrix(G)
S,V=eigh(L.toarray())
print(S)
print(L.toarray())
```
