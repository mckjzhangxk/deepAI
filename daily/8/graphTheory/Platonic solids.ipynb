```python
import networkx as nx
import numpy as np
```

```python
def regular_solid():
    G=nx.Graph()
    G.add_edge(1,2)
    G.add_edge(1,5)
    G.add_edge(1,6)
    
    G.add_edge(2,3)
    G.add_edge(2,8)
    
    G.add_edge(3,4)
    G.add_edge(3,10)
    
    G.add_edge(4,5)
    G.add_edge(4,12)
    
    G.add_edge(5,14)
    
    G.add_edge(6,7)
    G.add_edge(6,15)
    
    G.add_edge(7,8)
    G.add_edge(7,17)
    
    G.add_edge(8,9)
    
    G.add_edge(9,10)
    G.add_edge(9,18)
    
    G.add_edge(10,11)
    
    
    G.add_edge(11,12)
    G.add_edge(11,19)
    
    G.add_edge(12,13)
    
    G.add_edge(13,14)
    G.add_edge(13,20)
    
    G.add_edge(14,15)
    
    G.add_edge(15,16)
    
    G.add_edge(16,17)
    G.add_edge(16,20)
    
    G.add_edge(17,18)
    
    G.add_edge(18,19)
    
    G.add_edge(19,20)
    return G
def graph2Matrix(G):
    n=len(G.nodes)
    A=np.zeros((n,n))
    for nodeIdx,adj in G.adjacency():
        for adjIdx,w in adj.items():
            s,t= (nodeIdx-1,adjIdx-1)   
            if s!=t:
                A[s,t]=1 #w['weight'] if 'weight' in w else 1
    return A
def LaplacianMatrix(A):
    D=np.diag(A.sum(axis=1))
    return D-A
```

```python
G=regular_solid()
A=graph2Matrix(G)
L=LaplacianMatrix(A)
```

```python
from scipy.linalg import eigh
import matplotlib.pyplot as plt
%matplotlib inline
w, v = eigh(L) # w = eigenvalues, v = eigenvectors
n
x = v[:,1] 
y = v[:,2]
spectral_coordinates = {i+1 : (x[i], y[i]) for i in range(n)}


nx.draw(G,pos=spectral_coordinates)
nx.draw_networkx_labels(G,pos=spectral_coordinates)
plt.axis('equal')
```
