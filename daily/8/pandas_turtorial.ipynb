# A real Example(Abalone)
 <img src='images/abalone.jpg' width='480'/> <br />
* ### Predicting the age of abalone from physical measurements. The age of abalone is determined by cutting the shell through the cone, staining it, and counting the number of rings through a microscope -- a boring and time-consuming task. Other measurements, which are easier to obtain, are used to predict the age<br >

#### * Name / Data Type / Measurement Unit / Description 
----------------------------- 
* Sex / nominal / -- / M, F, and I (infant) 
* Length / continuous / mm / Longest shell measurement 
* Diameter	/ continuous / mm / perpendicular to length 
* Height / continuous / mm / with meat in shell 
* Whole weight / continuous / grams / whole abalone 
* Shucked weight / continuous	/ grams / weight of meat 
* Viscera weight / continuous / grams / gut weight (after bleeding) 
* Shell weight / continuous / grams / after being dried 
* Rings / integer / -- / +1.5 gives the age in years 


```python
abalone=pd.read_table('panda_dataset/abalone.data',sep=',',names=['sex','length','diameter','height',
                                                                  'weight','shuckled_weight',
                                                                  'viscera_weight','shell weight','rings'])
print(abalone.shape)
abalone.head(10)
```

```python
abalone.describe()
```

## How do I sort a pandas DataFrame or a Series?

```python
abalone.sort_values(['rings','length'],ascending=True)
# abalone['rings'].sort_values(ascending=False)
```

## How do I filter rows of a pandas DataFrame by column value?
* I want to see how the attribute distribution for all abalone's ring equal 12

```python
abalone[(abalone.rings>12) & (abalone.rings<=14)  ].sort_values('rings')
```

# How To Plot?

```python
mouth=np.arange(1,13)
price=mouth*3
support=10-mouth

db=np.stack([mouth,price,support],axis=1)
np.random.shuffle(db)
pd_db=pd.DataFrame(db,columns=['mouth','price','support'])
```

```python
pd_db.plot(x='mouth')
```

```python
abalone.plot(x='length',y='rings',kind='scatter')
```
