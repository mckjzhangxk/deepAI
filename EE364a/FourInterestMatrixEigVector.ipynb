$sin(\theta(j-1)+\varphi)+sin(\theta(j+1)+\varphi)=sin(\theta j+\varphi)2cos(\theta)$

可以知道,$当x_n={sin(\theta j+\varphi)}$的时候,满足$x_{j-1}-2x_{j}+x_{j+1}=(2cos(\theta)-2)x_j$
$$特征值:\lambda=(2-2cos(\theta))$$
$$特征向量:x_j={sin(\theta j+\varphi)}$$

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
sin_k_pi_x=lambda k,x: np.sin(k*np.pi*x)
cos_k_pi_x=lambda k,x: np.cos(k*np.pi*x)
cos_k_pi_plus_half_pi_x=lambda k,x: np.cos((k+0.5)*np.pi*x)
```

```python
def showResult(f,n,n_end,k=[1,2],title='label',resolution=1000,pad=0.1):
    '''
    f:采样的基函数
    n:是采样点,numpy array
    n:end:采样的2个端点,numpy array只要2个元素
    k:传递给f的参数,对应与第k个特征值的采样函数.
    
    '''
    plt.figure(figsize=(8,8))
    t=np.linspace(n_end[0]-pad,n_end[1]+pad,resolution)
    
    for _k in k:
        yt=f(_k,t)
        yn=f(_k,n)
        yend=f(_k,n_end)
        plt.plot(t,yt,'-',label='k=%d'%_k)
        plt.plot(n,yn,'yo')
        plt.plot(n_end,yend,'ro')
     
    plt.title(title)
    plt.xlabel('t(s(j))')
    plt.ylabel('f(t)')
    plt.legend()
```

$特征向量:v_k=\begin{pmatrix}v_k[0],v_k[1],\cdots,v_k[N]\end{pmatrix}$

$离散索引:j=\begin{pmatrix}1,2,\cdots,N\end{pmatrix}$

$离散索引端点:jend=\begin{pmatrix}0,N+1\end{pmatrix}$

$采样函数:t=s(j)$,表示把离散索引j转换到连续函数的区间上的N个点t

$v_k=f_k(t)$,表示第k个特征向量,来自$f_k$上的采样点t,$f_k$决定每个特征向量和特征值.

```python
#准备一下全局变量,特征值v(k)=(x1,....xN)_k对应函数f_,k(s(j))采样点
N=9
j=np.arange(1,N+1)
jend=np.array([0,N+1])
```

$$K_{3,3} = 
 \begin{pmatrix}
    2  & -1 &0  \\
    -1 & 2  &-1 \\
     0 & -1 &2 \\
 \end{pmatrix}$$
 上面矩阵的特征对2个端点要求$x_0=x_{N+1}=0$,所以它的特征值是$sin(k \pi t)上的离散点,t=\begin{pmatrix}\frac{1}{N+1} &\frac{2}{N+1}&\cdots&\frac{N}{N+1}\end{pmatrix}$,相隔样本点间距$h=\frac{1}{N+1}.$
 $$\theta(k)=\frac{k\pi}{N+1}$$
 $$\lambda(k)=2-2cos(\frac{k\pi}{N+1})$$

```python
sample_func=lambda index:index/(N+1)
fk=sin_k_pi_x

samples=sample_func(j)
sample_end=sample_func(jend)
showResult(fk,
            n=samples,
            n_end=sample_end,
            title='$f(t)=sin(k\pi t)$,$h=1/(N+1)$,N=%d,$x_j=f(jh)$'%N)
```

$$B_{3,3} = 
 \begin{pmatrix}
    1  & -1 &0  \\
    -1 & 2  &-1 \\
     0 & -1 &1 \\
 \end{pmatrix}$$
 上面矩阵的特征对2个端点要求$x_0=x_1,x_{N+1}=x_{N}$,所以它的特征值是$cos(k \pi t)上的离散点,t=\begin{pmatrix}\frac{1-0.5}{N} &\frac{2-0.5}{N}&\cdots&\frac{N-0.5}{N}\end{pmatrix}$,相隔样本点间距$h=\frac{1}{N}.$
 
  $$\theta(k)=\frac{k\pi}{N}$$
 $$\lambda(k)=2-2cos(\frac{k\pi}{N})$$

```python
fk=cos_k_pi_x
sample_func=lambda index:(index-0.5)/N

samples=sample_func(j)
samples_end=sample_func(jend)

showResult(fk,
           n=samples,
           n_end=samples_end,
           title='$f(t)=cos(k\pi t)$,$h=1/N,x_j=f((j-0.5)h)$,N=%d'%N)
```

$$T_{3,3} = 
 \begin{pmatrix}
    1  & -1 &0  \\
    -1 & 2  &-1 \\
     0 & -1 &2 \\
 \end{pmatrix}$$
 上面矩阵的特征对2个端点要求$x_0=x_1,x_{N+1}=0$,所以它的特征值是$cos(k \pi t)上的离散点,t=\begin{pmatrix}\frac{1-0.5}{N+0.5} &\frac{2-0.5}{N+0.5}&\cdots&\frac{N-0.5}{N+0.5}\end{pmatrix}$,相隔样本点间距$h=\frac{1}{N}.$
 
  $$\theta(k)=\frac{k\pi}{N+0.5}$$
 $$\lambda(k)=2-2cos(\frac{k\pi}{N+0.5})$$

```python
fk=cos_k_pi_plus_half_pi_x
sample_func=lambda index:(index-0.5)/(N+0.5)

samples=sample_func(j)
samples_end=sample_func(jend)

showResult(fk,
           n=samples,
           n_end=samples_end,
           title='$f(t)=cos((k\pi+0.5\pi) t)$,$h=1/(N+0.5),x_j=f((j-0.5)*h)$,N=%d'%N)
```
