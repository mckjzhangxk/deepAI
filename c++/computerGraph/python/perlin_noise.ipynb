```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1. random gradient
N=10
grad=np.random.randn(N)
```

```python
#define a interporlate function
def interp(g1,g2,num=30):
    dx=np.linspace(0,1,num)
    forward=dx*g1
    backward=(dx-1)*g2
    
    w2=3*(dx**2)-2*(dx**3)
    w1=3*((1-dx)**2)-2*((1-dx)**3)
    
    return w1*forward+w2*backward
```

```python
plt.figure(figsize=(10,10))
ret=[]
for i in range(N-1):
    value=interp(grad[i],grad[i+1])
    ret.append(value)
ret=np.concatenate(ret,0)
for x,g in enumerate(grad):
    plt.arrow(x,0,g/3,0,width=0.01,color=(0,1,0))
plt.plot(np.linspace(0,N-1,len(ret)),ret)
_=plt.xticks(np.linspace(0,N-1,N))
```
