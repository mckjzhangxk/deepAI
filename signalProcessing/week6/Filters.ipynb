```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

$$y[n]=\lambda y[n-1]+(1-\lambda)x[n]$$

$$H(z)=\frac{1-\lambda}{1-\lambda z^{-1}}$$

$$DTFT\{x[n]\}=\frac{1-\lambda}{1-\lambda e^{-j\omega}}$$

```python
omega=np.linspace(-np.pi,np.pi,100)
```

```python
def show(X,Y,title=''):
    plt.figure()
    plt.plot(X,Y)
    plt.title(title)
```

```python
for i in range(0,10):
    lm=0.6+i*0.04
    fn=(1-lm)/(1-lm*np.exp(-1j*omega))
    fn_abs=np.abs(fn)
    fn_phase=np.arctan2(fn.imag,fn.real)
    show(omega,fn_abs,'$\lambda$=%f'%(lm))
    show(omega,fn_phase,'$\lambda$=%f'%(lm))
```

### resonator(检测某些频率)

$$Pole1=\lambda e^{j \omega_0}$$
$$Pole2=\lambda e^{-j \omega_0}$$

$$H(z)=\frac{1-\lambda}{(1-p1 z^{-1})(1-p2 z^{-1})}$$

```python
w0=2
for i in range(0,5):
    lm=0.6+i*0.08
    p1=lm*np.exp(1j*w0)
    p2=np.conj(p1)
    
    fn=(1-lm)/(1-p1*np.exp(-1j*omega))/(1-p2*np.exp(-1j*omega))
    fn_abs=np.abs(fn)
    fn_phase=np.arctan2(fn.imag,fn.real)
    show(omega,fn_abs,'$\lambda$=%f'%(lm))
#     show(omega,fn_phase,'$\lambda$=%f'%(lm))
```

### DC removal

$$\sum {x[n]}=0$$

$$X(e^{j\omega})=\sum {x[n]e^{-j0n}}=0$$
$$T-transform(1)=0,have zoers=1$$

$$H(z)=1-z^{-1}$$
$$H(w)=1-e^{-jw}$$

```python
Hw=1-np.exp(-1j*omega)
show(omega,np.abs(Hw))
```

```python
lm=0.8
pole=lm
exp=np.exp(-1j*omega)
fn=(1-exp)/(1-pole*exp)
fn_abs=np.abs(fn)
fn_phase=np.arctan2(fn.imag,fn.real)
show(omega,fn_abs)
```

## Ham(remove some frequence)

$$\frac{(1-e^{j\omega}z^{-1})(1-e^{-j\omega}z^{-1})}{(1-\lambda e^{j \omega }z^{-1})(1-\lambda e^{-j \omega }z^{-1})}$$

```python
lm=0.99
w0=1

Z=np.exp(1j*w0)
P=lm*np.exp(1j*w0)

exp=np.exp(-1j*omega)
fn=(1-Z*exp)*(1-Z.conj()*exp)/(1-P*exp)/(1-P.conj()*exp)

fn_abs=np.abs(fn)
fn_phase=np.arctan2(fn.imag,fn.real)
show(omega,fn_abs)
```
